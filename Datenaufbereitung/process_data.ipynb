```python
import pandas as pd
import os
```

```python
path = "raw/"
file_type = ".json"

pdf = None

for filename in os.listdir(path=path):
    if filename.endswith(file_type):
        print(path+filename)
        df = pd.read_json(path+filename, encoding='ISO-8859-1')
        if pdf is None:
            pdf = df
        else:
            pdf = pd.concat([pdf,df])
```

```python
f_pdf = pdf.dropna()
f_pdf = f_pdf.drop_duplicates(subset=["id"])
f_pdf = f_pdf.set_index("id")
```

```python
display(f_pdf)
```

```python
f_pdf.to_pickle(path="export.pkl")
p_pdf = pd.read_pickle("export.pkl")
display(p_pdf)
```
